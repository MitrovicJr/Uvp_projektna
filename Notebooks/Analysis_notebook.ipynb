For the data analysis and visualization, we will need Pandas for handling the tabular data and matplotlib as well as seaborn for the graphs.

```python
#Importing pandas for data analysis and Matplotlib and Seaborn for data visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's load the data from the csv file that we extracted to the 'results' folder

```python
#Loading in the dataframe from the csv file, produced by the scraping.py file and contained the 'results' folder
df = pd.read_csv('../results/result_data.csv')
```

We can have a look at the first 10 rows of the dataframe using the .head() method

```python
df.head(10)
```

It looks like the columns 'Place' and 'Time', as well as 'Year of birth' need to be saved as an integer or float dtype, so let's check if that is done already

```python
#Displays relevant Dtype information for the dataframe, as well as checks the Null value count of the columns
df.info()
```

Clearly, we have to convert the 'Time' and 'Place' columns to the respective dtypes we want. We can do so for instance using the pd.to_numeric() method

```python
#Casting the 'Time' and 'Place' columns as float values
df['Time'] = pd.to_numeric(df['Time'], errors='coerce', downcast='float')
df['Place'] = pd.to_numeric(df['Place'], errors='coerce', downcast='float')
```

Now we should have te correct dtypes:

```python
df.dtypes
```

Now, we can begin to have a look at some of the data that could be interesting.

For instance, we can have a look at some of the most common birth years of the competing athletes along all of the dataframe:

```python
df['Year of birth'].value_counts()
```

It looks like the birthyears are somewhat centered around 1997 with some outliers.

As swimming is a sport, where young athletes tend to accomplish most, this makes quite a bit of sense. It is not uncommon to see swimmers retire at age 30 or even earlier, though short distance disciplines do allow for older swimmers to be very competitive as well.

We can have a brief look at the distribution of the data using the Pandas .describe() method, which provides us with the mean, standard deviation and quartiles, as well as a max and min value for the data

```python
#Quick statistical description of the 'Year of birth' column
df['Year of birth'].describe()
```

Looks like the hypothesis holds quite well. As we can see, the youngest athlete in the competition was born in 2005, and the oldest in 1983, a 22 year difference between them!

We can also graph the data to get a more intuitive sense of the distribution

```python
#Plotting the 'Year of birth' column as a histogram using seaborn
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 5))
sns.histplot(df['Year of birth'], bins=20, kde=False, color='skyblue', edgecolor='black')

plt.xlabel('Year', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title('Histogram of Years ', fontsize=16, fontweight='bold')
plt.yticks(fontsize=12)

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

As the dataframe contains all the data from the swimming events, it might be wise to create sub-dataframes of the preliminary rounds, semifinals and finals for further analysis, or comparison

```python
#Initiates a dictionary for filtering using the 'Event'column
keywords = {
    'Final': 'Final',
    'Semifinal': 'Semifinal',
    'Prelim': 'Prelim'
}

#Creates 3 dataframes with the relevant data by filtering on the 'Event' column
df_final = df[df['Event'].str.contains(keywords['Final'], case=True, na=False)]
df_semifinal = df[df['Event'].str.contains(keywords['Semifinal'], case=False, na=False)]
df_prelim = df[df['Event'].str.contains(keywords['Prelim'], case=False, na=False)]
```

Now lets create some side-by-side histograms and compare them to the overall distribution, to see how the ages of the athletes differ between the progressive stages of the competition

```python
#Plotting the 'Year of birth' histograms as before, this time using the extracted 'Final', 'Semifinal' and 'Prelim' dataframes
sns.set_theme(style="whitegrid")

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.histplot(df_final['Year of birth'], bins=20, kde=False, color='lightcoral', edgecolor='black')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Years - Final', fontsize=14, fontweight='bold')
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

plt.subplot(1, 3, 2)
sns.histplot(df_semifinal['Year of birth'], bins=20, kde=False, color='lightgreen', edgecolor='black')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Years - Semifinal', fontsize=14, fontweight='bold')
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

plt.subplot(1, 3, 3)
sns.histplot(df_prelim['Year of birth'], bins=20, kde=False, color='skyblue', edgecolor='black')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Histogram of Years - Prelim', fontsize=14, fontweight='bold')
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()


plt.show()
```

As we can see, the Final and semifinal data histograms seem to be quite similar, though they differ from the preliminary data. This means that the semifinals and finals seem to feature more athletes born around 2000 (proportionally) than the preliminary rounds of the competition.

However, first let's take a look at some of the countries that were the most represented in the swimming section of the olympics:

```python
#Creates a series with the most represented countries in the events
country_series = df.groupby('Country')['Name'].count().sort_values(ascending=False)

country_series
```

```python
#Plotting the top 10 most represented countries using a seaborn bar plot
top_10_countries = country_series.head(10)


plt.figure(figsize=(12, 5))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values, palette='crest')


plt.xlabel('Country', fontsize=14)
plt.ylabel('Representation', fontsize=14)
plt.title('Top 10 Countries by representation', fontsize=16, fontweight='bold')
plt.xticks(rotation=30, ha='center', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

We can see that USA and Russia sent the largest cohort of swimmers to the olympics, which might indicate that their chances of winning certain events will have been greater.

Let's examine this claim further by taking a look at the average placement of athletes in the finals, respective to their country

```python
#Showcases the Pandas series of average placement for each country in the finals of every event
placement_by_country = df_final.groupby('Country')['Place'].mean().sort_values(ascending=True)
placement_by_country
```

```python
#Graph for the previously created Pandas series on average swimmer placement in the finals
top_10_placement = placement_by_country.head(10)


plt.figure(figsize=(12, 5))
sns.barplot(x=top_10_placement.index, y=top_10_placement.values, palette='magma')


plt.xlabel('Country', fontsize=14)
plt.ylabel('Average Placement', fontsize=14)
plt.title('Top 10 Countries by Average Placement', fontsize=16, fontweight='bold')
plt.xticks(rotation=30, ha='center', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

We cna see that Russia is not among the top performers, which means that their athletes (on average) tended to not reach the finals or performed poorly in them. Tunisia and Finland seem to perform exceptionally well, but this could also be due to the very small amount of swimmers that competed from the countries.

Speaking of, we can see which athletes were most commonly seen at the swimming events, meaning that they were entered in the most evennts. In swimming of course, athletes can enter more than just 1 event anad can compete in additional relay and mixed relay events (2 men and 2 women per team)

```python
#Pandas series for the most events envtered by certain athletes 
events_per_athlete = df.groupby('Name')['Event'].count().sort_values(ascending=False)
events_per_athlete
```

```python
#Graphing the series
most_events = events_per_athlete.head(10)


plt.figure(figsize=(12, 5))
sns.barplot(x=most_events.index, y=most_events.values, palette='Blues')


plt.xlabel('Swimmer name', fontsize=14)
plt.ylabel('Amount of events entered in', fontsize=14)
plt.title('Top 10 Athletes with most events', fontsize=16, fontweight='bold')
plt.xticks(rotation=30, ha='center', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

Clearly, Caeleb Dressel was entered in the most events, followed closely by his USA counterpart Michael Andrew. Dressel had an exceptional olympics and one of the best swimming performances since the great mr. Phelps, Michael Andrew not so much. Though he was entered in many events, he did not achieve comperable success. This is also due to Andrew being an "all round swimmer", competing in many different strokes and not only 2 or 1 like usually.

Finally, lets take a look at which countries won the most at the olympics. We can do so by examining the number of gold medals per country:

```python
#Creates a series for the disribution of the medals at the olympics
df_most_gold = df_final[df_final['Place'] == 1].groupby('Country code')['Event'].count().sort_values(ascending=False)
df_most_gold
```

```python
#Plotting the medal distribution using a pie chart
plt.figure(figsize=(10, 6))

colors = sns.color_palette('crest', n_colors=len(df_most_gold))

plt.pie(
    df_most_gold,
    labels=df_most_gold.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)

plt.title('Distribution of Gold Medals by Country')

plt.show()
```

It looks like team USA had a very strong showing in the swimming events, with team Great Britain (largely due to the success of Adam Peaty) and Russia taking second place in the gold medal tally.
